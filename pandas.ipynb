```python
import pandas as pd
import numpy as np
```

```python
# Series
arr = np.array([10,20,30,40,50])
marks = pd.Series(arr)
print(marks)
print(marks[0])
```

```python
marks = pd.Series(
    [10,20,30],
    index = ["math","phy","chem"]
)
print(marks)
```

```python
marks = pd.Series(
    [10,20,30],
    index = ["math","phy","chem"]
)
print(marks)
print(marks["math"])
print(marks.shape)
print(marks.size)
print(marks.dtype)
print(marks.values)
print(marks.index)
```

```python
numbers =[1,2,34,7]
result = []
for i in numbers:
  result.append(i*2)
print(result)
```
